```python
import cv2
import win32gui
import win32con


eye_cascade = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')

cap = cv2.VideoCapture(0)

while 1:

    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = eye_cascade.detectMultiScale(gray, 1.3, 2)

    for (x,y,w,h) in faces:
        roi_gray = gray[y:1, x:1]
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)

    cv2.imshow('img', img)
    k = cv2.waitKey(20) & 0xff
    if k == 27:
        break
    if faces is ():
        win32gui.SendMessage(win32con.HWND_BROADCAST,
        win32con.WM_SYSCOMMAND, win32con.SC_MONITORPOWER, 2)
        print("Eye Contact Broken")
    elif faces is not ():
        win32gui.SendMessage(win32con.HWND_BROADCAST,
        win32con.WM_SYSCOMMAND, win32con.SC_MONITORPOWER, -1)
        print("***--Eye Contact Maintained--***")
        
cap.release()
cv2.destroyAllWindows()
```
